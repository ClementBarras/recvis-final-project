```python
split1_path = '../datasets/ucfTrainTestlist/trainlist01.txt'
split2_path = '../datasets/ucfTrainTestlist/trainlist02.txt'
split3_path = '../datasets/ucfTrainTestlist/trainlist03.txt'
```

```python
def read_file(path):
    with open(path) as f:
        lines = f.readlines()
        #lines = [line.split(' ')[0] for line in lines]
    f.close()
    return lines
```

```python
split1 = read_file(split1_path)
split2 = read_file(split2_path)
split3 = read_file(split3_path)
print(len(split1))
print(len(split2))
print(len(split3))
```

```python
import random
def split_train_val(split, path_train, path_val):
    random.shuffle(split)
    split_idx = int(len(split)*0.80)
    train_split = split[:split_idx]
    val_split = split[split_idx:]
    with open(path_train, 'w') as f:
        f.writelines(train_split)
        
    with open(path_val, 'w') as f:
        f.writelines(val_split)
```

```python
split_train_val(split1, '../datasets/ucfTrainTestlist/trainlist1.txt', '../datasets/ucfTrainTestlist/vallist1.txt')
split_train_val(split2, '../datasets/ucfTrainTestlist/trainlist2.txt', '../datasets/ucfTrainTestlist/vallist2.txt')
split_train_val(split3, '../datasets/ucfTrainTestlist/trainlist3.txt', '../datasets/ucfTrainTestlist/vallist3.txt')
```

```python
full_train = list(set(split1).union(set(split2)).union(set(split3)))
print(len(full_train))
split_train_val(full_train, '../datasets/ucfTrainTestlist/completetrainlist.txt', '../datasets/ucfTrainTestlist/completevallist.txt')
```

```python
from IPython.display import clear_output
from of_utils import of_slicing
import os

def add_max_of_slice(list_path, of_folder, width):
    with open(list_path) as f:
        lines = f.readlines()
    n_lines = len(lines)
    file_list = [line.split(" ")[0] for line in lines]
    class_list = [line.split(" ")[1] for line in lines]
    res = ""
    for i in range(n_lines):
        clear_output()
        file = file_list[i]
        file = file.split("/")[1].split(".")[0]
        classe = class_list[i][:-1]
        print("{} ({}/{})".format(file, i+1, n_lines))
        idx = of_slicing(of_folder, file, width = 100)
        res += "{},{},{}\n".format(file, classe, idx)
    return res
res = add_max_of_slice('../datasets/ucfTrainTestlist/completetrainlist.txt', '../datasets/tvl1_flow/average_flow/', 100)    
```

```python
with open('../datasets/ucfTrainTestlist/completetrainlist_flow.txt', "w") as f:
    f.write(res)
```
