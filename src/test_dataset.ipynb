```python
from dataset import ProxyTaskDataset
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import torch
```

```python
train_set = ProxyTaskDataset()
loader = DataLoader(train_set, batch_size=1, shuffle=True)
```

```python
a = next(iter(loader))
for i in a:
    fig, ax = plt.subplots(1,6, figsize=(14,14))
    for j, im in enumerate(i[0]):
        im = torch.squeeze(im)
        ax[j].imshow(im)
```
