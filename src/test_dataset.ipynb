```python
from dataset import ProxyTaskDataset
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import torch
```

```python
train_set = ProxyTaskDataset(sampling='constrained_consecutive')
loader = DataLoader(train_set, batch_size=2, shuffle=True)
```

```python
a = next(iter(loader))
for i in a:
    print(i[1])
    fig, ax = plt.subplots(1,6, figsize=(14,14))
    for j, im in enumerate(i[0]):
        im = torch.squeeze(im[0])
        ax[j].imshow(im)
```
